```python
import numpy as np
import pandas as pd
import ast
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
events = pd.read_excel('excel/events.xlsx')
types = pd.read_excel('excel/types.xlsx')
```

```python
events
```

```python
types
```

```python
events = pd.merge(events, types, left_on=  'type_id',
                   right_on= 'id')
events
```

```python
events = events.drop('type_id', axis=1)
events = events.drop('online', axis=1)
events = events.drop('id_y', axis=1)
events
```

```python
events = events.rename({'id_x': 'id', 'name': 'group'}, axis=1)
events
```

```python
events = events.sort_values('id')
events
```

```python
events['district'] = events['district'].apply(lambda x:x.split(';'))
events
```

```python
events
```

```python
# events.loc[7, ['district']] = list(set(events.loc[7]['district']))
events['district'] = [list(filter(None, list(set(ev_dis)))) for ev_dis in events['district']]
```

```python
events
```

```python
events['overwiev'] = events['overwiev'].apply(lambda x:x.split())
```

```python
events['group'] = [[ev] for ev in events['group']]
```

```python
events
```

```python
events['tags'] = events['overwiev'] + events['district'] + events['group']
```

```python
new = events.drop(columns=['overwiev','district','group'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
with open('stopwords.txt') as f:
    lines = f.read().splitlines()
lines
```

```python
cv = CountVectorizer(encoding='koi8r', stop_words=lines)
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'Флористика'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Шахматы')
```

```python
pickle.dump(new,open('pkl/event_list.pkl','wb'))
pickle.dump(similarity,open('pkl/similarity.pkl','wb'))
```
